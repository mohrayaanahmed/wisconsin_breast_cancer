```python
# IMPORT PANDAS TO READ THE DATASET

import pandas as pd
data = pd.read_csv("wdbc.csv") # READ THE DATASET
data
```

```python
# Find the total number of attributes

data.count(axis = 1) # ANSWER: 32
```

```python
# Find the total number of malignant and benign cases

pd.crosstab(index = data["Diagnosis"], columns = "count")
# ANSWER: B-357, M-212
```

```python
# Find the Mean, Variance and STD for each real attribute

y = data.describe() # Calculate mean, std
del y["ID Number"] # DELETE THE FIRST ATTRIBUTE (ID)

y
```

```python
y.iloc[1] # Print the mean row
```

```python
y.iloc[2] # Print the Standard Deviation row
```

```python
for i in y.iloc[2]:
    print(i*i)  # Printing the Variance (Since Variance = (STD)^2)
```

```python
# Split the dataset on the basis of diagnosis (B or M)

total_data = data.groupby(data["Diagnosis"])
```

```python
#FOR BENIGN

data_benign = total_data.get_group("B")
temp = data_benign.copy()
del temp["ID Number"]
temp = temp.describe()
```

```python
print(temp.iloc[1]) # MEAN OF THE BENIGN DATASET
```

```python
print(temp.iloc[2]) # STANDARD DEVIATION OF BENIGN DATASET
```

```python
for i in temp.iloc[2]:
    print(i*i)  # PRINTING THE VARIANCE OF BENIGN DATASET
```

```python
#FOR MALIGNANT
data_malignant = total_data.get_group("M")
temp_2 = data_malignant.copy()
del temp_2["ID Number"]
temp_2 = temp_2.describe()
```

```python
print(temp_2.iloc[1]) # MEAN FOR THE MALIGNANT DATASET
```

```python
print(temp_2.iloc[2]) # STANDARD DEVIATION OF MALIGNANT DATASET
```

```python
for i in temp_2.iloc[2]:
    print(i*i)  # PRINTING THE VARIANCE OF MALIGNANT DATASET
```

```python
# DATA NORMALIZATION

x = data.copy() # COPY THE DATA INTO A VARIABLE x

# Delete non-numerical values
del x["ID Number"]
del x["Diagnosis"]

# Normalize to unit variance
data_normalized = (x - x.mean())/x.std()
```

```python
# INSERT THE COLUMNS (ID, DIAGNOSIS) BACK IN THE NORMALIZED DATASET
data_normalized.insert(0, column = "Diagnosis", value = data["Diagnosis"])

#data_normalized.to_csv("Normalized Data.csv") # Save in Excel for future use

total_data_normalized = data_normalized.groupby(data["Diagnosis"])
data_benign = total_data_normalized.get_group("B")
data_malignant = total_data_normalized.get_group("M")

data_normalized.head()
```
